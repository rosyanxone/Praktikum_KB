- Mengimport <b>library</b> yang akan digunakan dan diperlukan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- Membaca dataset dengan library pandas

```python

df = pd.read_csv("movies.csv")
df.info()
```

- Mengubah tipe data pada atribut <i>score</i> dari float menjadi bilangan bulat

```python
df["score"] = df["score"].astype("int64")
```

- Melakukan perbandingan pada atribut <b>Rating</b> dengan <b>Score</b> menggunakan grafik batang

```python
plt.figure(figsize = (15, 5))
sns.countplot(data=df, x='rating', palette='viridis', hue='score')
plt.title("Perbandingan Rating dan Score")
plt.xlabel('Rating')
plt.ylabel('Score')
plt.show()
```

- Menunjukan hubungan antara dua variable <b>Year</b> dan <b>Gross</b> dengan Scatterplot

```python
plt.figure(figsize = (15, 8))
sns.scatterplot(data = df, x = "year", y = "gross")
plt.title("Hubungan antar tahun dan gross pada Scatterplot")
plt.xlabel('year')
plt.ylabel('gross')
```

- Menampilkan pemetaan untuk menunjukkan korelasi antar variable pada dataset <i>movies</i>

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(), center=0)
plt.title("Pemetaan korelasi tiap atribut")
plt.xlabel('atribut x')
plt.ylabel('atribut y')
```

- Penggunaan grafik <b>Pie Chart</b> pada data <i>categorical</i> yaitu <b>score</b>

```python
data_pie = df["score"].value_counts().rename_axis("score").reset_index(name="nomor")
plt.figure(figsize=(12, 10))
plt.pie(data_pie["nomor"],labels=data_pie["score"], startangle=120, autopct='%.1f%%')
plt.title("Grafik Pie Score")
plt.xlabel('Score')
plt.ylabel('Kuantitas')
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=1)
plt.show()
```
