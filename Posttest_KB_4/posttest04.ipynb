- Melakukan <b>Data cleaning</b> pada data dengan <b>nilai null</b>. Mengganti nilai null sesuai ketentuan. (<b>bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus</b>).

```python
print("Jumlah record yang memiliki nilai null: ")
print(df.isna().sum())

tipeData_creditProduct = df.dtypes['Credit_Product']
print("\nTipe data atribut yg memiliki nilai null: ")
print(tipeData_creditProduct)
```

```python
from sklearn.impute import SimpleImputer
SimpleImputer = SimpleImputer(strategy='most_frequent' )
df["Credit_Product"] = SimpleImputer.fit_transform(df[["Credit_Product"]])

print("\nJumlah nilai null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```
